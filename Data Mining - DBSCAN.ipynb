## DBSCAN Cho một ảnh. 

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import cv2

# Đọc hình ảnh và chuyển đổi sang không gian màu LAB
image = cv2.imread('/content/drive/MyDrive/Colab Notebooks/DM - MAT3534/Data/Testing Data set/slide1.webp')
image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

plt.imshow(image)
plt.show()

# Chuẩn hóa dữ liệu
data = image.reshape((-1, 3))
data = StandardScaler().fit_transform(data)

# Phân cụm dữ liệu bằng DBSCAN
dbscan = DBSCAN(eps=0.01, min_samples=7)
dbscan.fit(data)
labels = dbscan.labels_
print(set(labels))

# Hiển thị kết quả phân đoạn
segmented_image = labels.reshape(image.shape[:2])
plt.imshow(segmented_image)
plt.show()
```

## KMeans cho phân đoạn ảnh. 

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import cv2

# Đọc hình ảnh và chuyển đổi sang không gian màu LAB
image = cv2.imread('Data\OralCancer\non-cancer\close-up-of-spots-on-tongue.jpg')
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

# Chuẩn hóa dữ liệu
data = image_lab.reshape((-1, 3))
scaler = StandardScaler()
data = scaler.fit_transform(data)

# Phân cụm dữ liệu bằng K-means
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
labels = kmeans.labels_

# Hiển thị kết quả phân đoạn
segmented_image = labels.reshape(image_lab.shape[:2])
plt.imshow(segmented_image)
plt.show()

kmeans.cluster_centers_
```

```python
segmented_image.shape
```

```python
kmeans.cluster_centers_
```

```python
centers = kmeans.cluster_centers_
centers = scaler.inverse_transform(centers)
centers = centers.reshape((1, -1, 3)).astype(np.uint8)
centers = cv2.cvtColor(centers, cv2.COLOR_LAB2BGR).reshape((-1, 3))

for i, center in enumerate(centers):
    # Vẽ các tâm cụm lên hình ảnh
    x, y = int(center[0]), int(center[1])
    cv2.circle(image, (x, y), 10, (0, 255, 0), -1)

    # Tạo một mặt nạ cho cụm hiện tại
    mask = labels.reshape(image.shape[:2]) == i
    mask = mask.astype(np.uint8) * 255

    # Tính toán tọa độ của các góc của hình chữ nhật bao quanh các điểm trong cụm
    x, y, w, h = cv2.boundingRect(mask)
    # print('w = ', w)
    # print('h = ', h)
    # h = 20
    # w = 20
    # Vẽ hình chữ nhật lên hình ảnh
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 1)

# Hiển thị hình ảnh
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
